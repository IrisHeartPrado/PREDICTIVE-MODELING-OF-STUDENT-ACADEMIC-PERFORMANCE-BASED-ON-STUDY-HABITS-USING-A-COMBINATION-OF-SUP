Independent variables are the predictor or the features that will help us predict whether the student has probability to be a irregular student for the next semester or will probably stay as regular student based on their study habits that is gathered through the survey and final grades

Dependent variable is the target variable or what we are trying to know which is the status of the student

```python
# Dependent(y) and Independent(x) Variable segregation 

x = concat_data_updated.drop(['Status'], axis=1)
y = concat_data_updated['Status']
```

```python
#Data Spliting (Training and Testing Set)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=25)
```

```python
import pickle
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report

# Load pre-trained base learners
with open('SVM_Solo_woGans_convert.pkl', 'rb') as f_svm:
    svm_model = pickle.load(f_svm)

with open('RANDOM FOREST_GANS_25 convert.pkl', 'rb') as f_rf:
    rf_model = pickle.load(f_rf)

# Define StackingClassifier with meta-learner
meta_learner = StackingClassifier(
    estimators=[('svm', svm_model), ('rf', rf_model)],
    final_estimator=LogisticRegression(solver='liblinear') 
)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define the parameter grid for Logistic Regression
param_grid = {
    'final_estimator__C': [ 0.1, 1, 2],  # Regularization parameter
    'final_estimator__penalty': ['l1', 'l2'],  # Regularization type
    'final_estimator__solver': ['liblinear', 'saga']  # Solver options
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(meta_learner, param_grid, cv=10,scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Display the best parameters
print("Best parameters:", grid_search.best_params_)

```

```python

# Perform Cross-Validation
cv_scores = cross_val_score(meta_learner, x, y, cv=5, scoring='accuracy')  # Adjust 'cv' as needed
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())
print("Standard Deviation of CV Score:", cv_scores.std())

# Refit the meta-learner on the full training data
meta_learner.fit(x_train, y_train)

# Access meta-learner coefficients (after fitting)
if hasattr(meta_learner.final_estimator_, 'coef_'):
    coefficients = meta_learner.final_estimator_.coef_
    for idx, (name, _) in enumerate(meta_learner.estimators):
        print(f"Weight of {name}: {coefficients[0][idx]}")
else:
    print("Meta-learner does not have coefficients available.")

    
# Evaluate the model with the best parameters
best_meta_learner = grid_search.best_estimator_
y_pred_best = best_meta_learner.predict(x_test)

# Performance evaluation
test_accuracy = accuracy_score(y_test, y_pred_best)
print("Test Accuracy with best parameters:", test_accuracy)
print("Classification Report with best parameters:")
print(classification_report(y_test, y_pred_best))
```
